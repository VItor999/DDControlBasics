```python
# -*- coding: utf-8 -*-
"""
Created on Sun Nov 17 12:25:31 2019
https://www.jcbrolabs.org/python-codes
@author: mm
"""
import os
from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg
from matplotlib.figure import Figure
import tkinter as tk
import numpy as np
import serial as sr
from tkinter.filedialog import askdirectory
# add figure canvas
fig = Figure();

#------global variables
data = np.array([])
u = np.array([])
k = np.array([],dtype = np.int_)
i = 0
cond = False
numAmostras = 3000
#-----plot data-----
def plot_data():
    global cond, data, numAmostras,i,k,u
    
    if (cond == True):
        
        a = s.readline()
        a = a.decode()
        str = a[:-2]
        str = str.split('#')
        i+=1
        if(len(data) < numAmostras):
            data = np.append(data,float(str[0])) # adicionar elementos conforme a necessidade
            k = np.append(k,i) 
            u = np.append(u,float(str[1]))
        else:
            data[0:numAmostras-1] = data[1:numAmostras]
            data[numAmostras-1] = float(str[0])
            u[0:numAmostras-1] = u[1:numAmostras]
            u[numAmostras-1] = float(str[1])
            k[0:numAmostras-1] = k[1:numAmostras]
            k[numAmostras-1] = i
        line1.set_xdata(np.arange(0,len(data)))
        line1.set_ydata(data)
        line2.set_xdata(np.arange(0,len(data)))
        line2.set_ydata(u)
        canvas.draw()
    
    root.after(1,plot_data)

def on_closing():
    global s
    s.close()
    root.destroy()

def save_data():
    global data,k
    img_dir = askdirectory()
    os.chdir(img_dir)
    salvar = np.column_stack((k,data,u))
    np.savetxt("teste.csv", salvar,fmt='%5d , %.6e , %6e ' )
    
def plot_start():
    global cond
    cond = True
    s.reset_input_buffer()

def plot_stop():
    global cond
    cond = False
def save_fig():
    global fig
    img_dir = askdirectory()
    os.chdir(img_dir)
    fig.savefig('grafico.png')
    

#-----Main GUI code-----
root = tk.Tk()
root.title('Referência e Saída')
root.configure(background = 'light blue')
root.geometry("525x500") # set the window size

#------create Plot object on GUI----------

ax = fig.add_subplot(111)

#ax = plt.axes(xlim=(0,100),ylim=(0, 120)); #displaying only 100 samples
ax.set_title('Referência e Saída')
ax.set_xlabel('Últimas Amostras')
ax.set_ylabel('Tensão(V)')
ax.yaxis.set_ticks(np.arange(0.0, 10.0, 1))
ax.set_xlim(0,numAmostras)
ax.set_ylim(-0.5,10.5)

line1 = ax.plot([],[],label = 'Referência')[0]
line2 = ax.plot([],[],label = 'Planta')[0] # adicionar linhas com o que quero
ax.legend(fancybox=True, framealpha=0.5,  borderpad=1,loc ='upper right')
#fig.legend(['a','b'])
ax.grid()
canvas = FigureCanvasTkAgg(fig, master=root)  # A tk.DrawingArea.
canvas.get_tk_widget().place(x = 10,y=10, width = 500,height = 400)
canvas.draw()

#----------create button---------
root.update();
start = tk.Button(root, text = "Iniciar", font = ('calbiri',12),command = lambda: plot_start())
start.place(x = 50, y = 450 )

root.update();
stop = tk.Button(root, text = "Parar", font = ('calbiri',12), command = lambda:plot_stop())
stop.place(x = 125, y = 450)

root.update();
stop = tk.Button(root, text = "Salvar Imagem", font = ('calbiri',12), command = lambda:save_fig())
stop.place(x = 200, y = 450)

root.update();
stop = tk.Button(root, text = "Salvar Dados", font = ('calbiri',12), command = lambda:save_data())
stop.place(x = 350, y = 450)

#----start serial port----
s = sr.Serial('COM7',9600);
s.reset_input_buffer()

root.resizable(False, False)
root.protocol("WM_DELETE_WINDOW", on_closing)
root.after(1,plot_data)
root.mainloop()
```
